```python
df2['date'] = df2['date'].astype('datetime64[ns]')
```

```python
# Extracting month and year from date to group by month and date for time series
df2['month_year'] = pd.to_datetime(df2['date']).dt.to_period('M')
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df2.dtypes
```

```python
# Create new df with aggregated values by month adn year
df2 = df2.groupby(['location', 'continent', 'iso_code', 'month_year', 'population'], as_index=False).agg(new_cases = ('new_cases', 'sum'), total_cases = ('total_cases', 'max'), new_deaths = ('new_deaths', 'sum'), total_deaths = ('total_deaths', 'max'))
```

```python
df2.head()
```

```python
df2.rename(columns = {'location' : 'country'}, inplace = True)
```

```python
df2['month_year'] = df2['month_year'].astype('str')
```

```python
df['month_year'] = df['month_year'].astype('str')
```

```python
new_df = df.merge(df2, on = ['country', 'iso_code', 'month_year'], how = 'outer', sort=False)
```
