## 1. Changing daily entries to monthly

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Create path to dataset
path = r'C:\Users\M de Villiers\Desktop\Data Analytics\Achievement 6\02_Data\Prepared_Data'
```

```python
# import data
df= pd.read_csv(os.path.join(path, 'covid_vaccinations_clean.csv'), index_col=False)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
# Extracting month and year from date to group by month and date for time series
df['month_year'] = pd.to_datetime(df['date']).dt.to_period('M')
```

```python
df.head()
```

```python
df.tail()
```

```python
# Export new dataset
df.to_csv(os.path.join(path, 'COVID-19_vaccinations_2.csv'), index=False)
```

```python
df1 = df.groupby(['country', 'iso_code', 'month_year', 'vaccines'], as_index=False).agg(total_vaccinations = ('total_vaccinations', 'max'), people_vaccinated = ('people_vaccinated', 'max'), people_fully_vaccinated = ('people_fully_vaccinated', 'max'), monthly_vaccinations_raw = ('daily_vaccinations_raw', 'sum'), monthly_vaccinations = ('daily_vaccinations', 'sum'), total_vaccinations_per_hundred = ('total_vaccinations_per_hundred', 'max'), people_vaccinated_per_hundred = ('people_vaccinated_per_hundred', 'max'), people_fully_vaccinated_per_hundred = ('people_fully_vaccinated_per_hundred', 'max'), monthly_vaccinations_per_million = ('daily_vaccinations_per_million', 'sum'))
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1.shape
```

```python
# Export monthly dataset
df1.to_csv(os.path.join(path, 'monthly_COVID-19_vaccinations.csv'), index=False)
```

# 3. Cleaning & Wrangling new dataset

```python
# dropping columns: Instead of dropping columns, I will just create a new dataset with the columns I need
# I want to drop more columns than I actually need so I think this way is easier
df3 = df2[['Country (or dependency)', 'Population (2020)']]
```

```python
df3.rename(columns = {'Country (or dependency)' : 'country'}, inplace=True)
```

# 4. Combining datasets

```python
df_merged = df3.merge(df1, on = 'country', how = 'outer', sort=False)
```
