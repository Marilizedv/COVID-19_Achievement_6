```python
df2['Date_reported'] = df2['Date_reported'].astype('datetime64[ns]')
```

```python
df2.head()
```

```python
# Extracting month and year from date to group by month and date for time series
df2['month_year'] = pd.to_datetime(df2['Date_reported']).dt.to_period('M')
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df2.dtypes
```

```python
# Create new df with aggregated values by month adn year
df2 = df2.groupby(['Country', 'Country_code', 'month_year'], as_index=False).agg(new_cases = ('New_cases', 'sum'), cumulative_cases = ('Cumulative_cases', 'max'), new_deaths = ('New_deaths', 'sum'), cumulative_deaths = ('Cumulative_deaths', 'max'))
```

```python
df2.head()
```

```python
df2.rename(columns = {'Country' : 'country'}, inplace = True)
```

```python
df2.rename(columns = {'Country_code' : 'country_code'}, inplace = True)
```

```python
df2['month_year'] = df2['month_year'].astype('str')
```

```python
df['month_year'] = df['month_year'].astype('str')
```

```python
new_df = df.merge(df2, on = ['country', 'month_year'], how = 'outer', sort=False)
```
